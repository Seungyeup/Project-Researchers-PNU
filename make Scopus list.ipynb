```python
import math
import pandas as pd
from pandas import DataFrame as df
from elsapy.elsclient import ElsClient
from elsapy.elsprofile import ElsAuthor, ElsAffil
from elsapy.elsdoc import FullDoc, AbsDoc
from elsapy.elssearch import ElsSearch
import json

# 정제조건
# 1. source type
src_type_list = ['Journal','Book','Book Series','Conference Proceeding']
# 2. document type
doc_type_list = ['Article','Book','Chapter','Conference Paper','Review','Business Article','Article in Press']

researchers = pd.read_excel("저자리스트(AUID 2020.7월 3일(최신)).xlsx")
scopus_all = df(columns=['대학','학과','전공','성명','교직원번호','PNU-ID','author name','AU-ID','year','Title','논문타입','EID','Source',
                        'prism:doi',
                        'dc:creator',
                        'prism:volume',
                        'prism:pageRange', 
                        'prism:coverDisplayDate', 
                        'citedby-count',
                        'subtype',
                        'subtypeDescription', 
                        'source-id', 
                        'openaccess',
                        'openaccessFlag'
                        ])

# api-key
# institue-key miss

con_file = open("config.json")
config = json.load(con_file)
con_file.close()

# key-client
client = ElsClient(config['apikey'])
```

```python
prev=0
for pro in range(0,len(researchers)):
    au_id_num=researchers['AU-ID'].loc[pro]
    name = researchers['성명'].loc[pro]
    gyo = researchers['교직원번호'].loc[pro]
    colle = researchers['대학'].loc[pro]
    maj = researchers['학과'].loc[pro]
    jun = researchers['전공'].loc[pro]
    pnu_id = researchers['PNU_ID'].loc[pro]
    en_name = researchers['풀네임'].loc[pro]
    
    if math.isnan(au_id_num):
        continue
    au_id = "au-id("+str(int(au_id_num))+")"
    tmp_search = ElsSearch(au_id,"scopus")
    tmp_search.execute(client)
    tmp_result = tmp_search.results_df
    result = tmp_result['dc:title']
    for doc_num in range(0,len(result)):
        # 정제조건 0, 년도는 2014 이후
        nonmun_year = tmp_result['prism:coverDate'].loc[doc_num].year
        if nonmun_year < 2014:
            continue
        if nonmun_year == 2020:
            continue
        
        # 정제조건 1, source type
        src_type = tmp_result['prism:aggregationType'].loc[doc_num]
        if not src_type in src_type_list:
            continue
            
        # 정제조건 2, document type
        doc_type = tmp_result['subtypeDescription'].loc[doc_num]
        if not doc_type in doc_type_list:
            continue
        scopus_all.loc[prev] = [
                                        colle,
                                        maj,
                                        jun,
                                        name,
                                        gyo,
                                        pnu_id,
                                        en_name,
                                        au_id,
                                        nonmun_year,
                                        tmp_result['dc:title'].loc[doc_num],
                                        src_type,
                                        tmp_result['eid'].loc[doc_num],
                                        tmp_result['prism:publicationName'].loc[doc_num],
            # after database set
            tmp_result['prism:doi'].loc[doc_num],
            tmp_result['dc:creator'].loc[doc_num],
            tmp_result['prism:volume'].loc[doc_num],
            tmp_result['prism:pageRange'].loc[doc_num],
            tmp_result['prism:coverDisplayDate'].loc[doc_num],
            tmp_result['citedby-count'].loc[doc_num],
            tmp_result['subtype'].loc[doc_num],
            doc_type,
            tmp_result['source-id'].loc[doc_num],
            tmp_result['openaccess'].loc[doc_num],
            tmp_result['openaccessFlag'].loc[doc_num]
                                  ]
        prev += 1
    print(name)
    
```

```python
scopus_all.to_excel("DB/SCOPUS/scopus_all.xlsx")
```
