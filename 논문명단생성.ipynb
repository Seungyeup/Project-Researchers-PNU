```python
import pandas as pd
import numpy as np
from pandas import DataFrame as df
from openpyxl import load_workbook

pip_list = pd.read_excel("PIP_ALL.xlsx")
scopus_list = pd.read_excel("temp_scopus_all.xlsx")
researcher_list = pd.read_excel("저자test1.xlsx")

#key는 저자명 -> PIP 순회, SCOPUS순회

nonmun_list = df(columns=['대학','학과','전공','성명','풀네임','년도','논문명','저널명','Status','J-Status','J-DB','DT-P','DT-S','DOI'])
```

```python
num_cnt=0
for res in researcher_list['성명'].tolist():
    for pip_num in range(0,len(pip_list)):      # 먼저 PIP 기준으로 검색하는데
        if pip_list['성명'].loc[pip_num]==res:  # 해당 ROW성명이 저자와 일치한다면
            # 정보 일단 당겨오기
            nonmun_list.loc[num_cnt] = [ 
                pip_list['대학'].loc[pip_num], # 대학
                pip_list['학과'].loc[pip_num], # 학과
                pip_list['전공'].loc[pip_num], # 전공
                pip_list['성명'].loc[pip_num], # 성명
                None, #pip_list['대학'].loc[pip_num], # 풀네임
                None, #pip_list['대학'].loc[pip_num], # 년도
                pip_list['제목'].loc[pip_num], # 논문명
                None, #pip_list['대학'].loc[pip_num], # 저널명
                "P-only", # Status
                None, #pip_list['대학'].loc[pip_num], # J-Status
                None, #pip_list['대학'].loc[pip_num], # J-DB
                None, #pip_list['대학'].loc[pip_num], # DT-P
                None, #pip_list['대학'].loc[pip_num], # DT-S
                None, #pip_list['대학'].loc[pip_num]  # DOI
            ]
            if pip_list['제목'].loc[pip_num] in scopus_list['논문제목'].tolist():
                nonmun_list['Status'].loc[num_cnt] = "Both"
        num_cnt += 1
    for scopus_num in range(0,len(scopus_list)):
        if scopus_list['성명'].loc[scopus_num]==res:
            if scopus_list['논문제목'].loc[scopus_num] not in nonmun_list['논문명'].tolist():
                nonmun_list.loc[num_cnt] = [ 
                None, #pip_list['대학'].loc[pip_num], # 대학
                None, #pip_list['학과'].loc[pip_num], # 학과
                None, #pip_list['전공'].loc[pip_num], # 전공
                scopus_list['성명'].loc[scopus_num], # 성명
                None, #pip_list['대학'].loc[pip_num], # 풀네임
                None, #pip_list['대학'].loc[pip_num], # 년도
                scopus_list['논문제목'].loc[scopus_num], # 논문명
                None, #pip_list['대학'].loc[pip_num], # 저널명
                "S-only", # Status
                None, #pip_list['대학'].loc[pip_num], # J-Status
                None, #pip_list['대학'].loc[pip_num], # J-DB
                None, #pip_list['대학'].loc[pip_num], # DT-P
                None, #pip_list['대학'].loc[pip_num], # DT-S
                None, #pip_list['대학'].loc[pip_num]  # DOI
            ]
        num_cnt += 1 
                
```

```python
nonmun_list.reset_index(drop=True).to_excel("research_all.xlsx")
```
